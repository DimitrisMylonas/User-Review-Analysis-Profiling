```python
df = pd.read_csv('reviews.csv')
df.head()
```

```python
from nltk import word_tokenize

df['preprocessed_text'] = df['text'].str.lower()
df['preprocessed_text'] = df['preprocessed_text'].apply(word_tokenize)
df
```

```python
from string import punctuation

words_without_punctuation=[]

for index, row in df.iterrows():
    words_without_punctuation.append([word for word in row['preprocessed_text'] if word not in punctuation])

df['preprocessed_text'] = words_without_punctuation
df
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
words_without_stop_words=[]

for index, row in df.iterrows():
    words_without_stop_words.append([word for word in row['preprocessed_text'] if word not in stop_words])

df['preprocessed_text'] = words_without_stop_words
df
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('words')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words=[]

for index, row in df.iterrows():
    lemmatized_words.append([lemmatizer.lemmatize(word) for word in row['preprocessed_text']])

df['preprocessed_text'] = lemmatized_words
df
```

```python
from nltk.corpus import words

english_words = words.words()
english_words_only=[]

for index, row in df.iterrows():
    english_words_only.append([word for word in row['preprocessed_text'] if word in english_words])

df['preprocessed_text'] = english_words_only
df
```

```python
new_preprocessed_text=[]

for index, row in df.iterrows():
    new_preprocessed_text.append(','.join(row['preprocessed_text']))

df['preprocessed_text'] = new_preprocessed_text
df
```

```python
import csv

df.to_csv("preprocessed_reviews.csv", index=False, encoding='utf-8-sig', quoting=csv.QUOTE_ALL)
```
